```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
data = np.genfromtxt('movieReplicationSet.csv', delimiter = ',',skip_header = 1)
```

```python
import pandas as pd
df = pd.read_csv('movieReplicationSet.csv',skipinitialspace=True)
df= df.rename(columns={'Rambo: First Blood Part II': 'Rambo: First Blood Part II (1985)'})
titles = df.columns[0:400]
titles = titles.values.tolist()
```

```python
tyears = [[]]
index= []
x,y = data.shape
for i in range(400):
    if titles[i][-5:-1:1].isnumeric():
        years = int(titles[i][-5:-1:1])
        index.append(i)
        tyears[0].extend([years])
print(np.median(tyears[0]))
yandi = zip(tyears[0],index)
yandi = list(yandi)
```

```python
hyears= []
lyears= []
for i in range(len(yandi)):
    if yandi[i][0]<=np.median(tyears[0]) and len(lyears)< 200:
        lyears.append(yandi[i])
    else:
        hyears.append(yandi[i])
```
